# Meta Interview Practice Question

```python
friends = [[1,3], [2,3], [3,5], [4]]
```

```python
for x in friends:
    print(x)
```

```python
for x in friends:
    print(len(x))
```

```python
for x in friends:
    if len(x) == 1: print('no friends')
    elif len(x) > 1: print('has friends')
```

```python
dictionary = {}
for x in friends:
    if len(x) == 1:
        dictionary[x[0]] = 0
    else:
        dictionary[x[0]] = 1
        dictionary[x[1]] = 1
        
dictionary
```

```python
dictionary = {}
for x in friends:
    dictionary[x[0]] = 0
    if len(x) > 1:
        dictionary[x[0]] += 1
        dictionary[x[1]] = 1
        
dictionary
```
